```python
#q1
## Escreva uma função que receba uma lista de números e retorne outra lista com os números ímpares.

def coletar_numeros():
    qtd = int(input("quantos números você vai colocar? "))
    lista = []
    impar = []
    i = 1

    while i <= qtd:
        numero = int(input("insira um número na lista: "))
        lista.append(numero)
        if numero % 2 == 1:
            impar.append(numero)
        i += 1

    print("lista original:", lista)
    print("lista impares:", impar)
    return lista, impar

coletar_numeros()
```

```python
#q2
## Escreva uma função que receba uma lista de números e retorne outra lista com os números primos presentes

def eh_primo(n):
    if n < 2:
        return False
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True


def coletar_numeros():
    qtd = int(input("quantos numeros você vai colocar? "))
    lista = []
    primo = []
    i = 1

    while i <= qtd:
        numero = int(input("insira um número na lista: "))
        lista.append(numero)
        if eh_primo(numero):
            primo.append(numero)
        i += 1

    print("lista original:", lista)
    print("lista primos:", primo)
    return lista, primo

coletar_numeros()
```

```python
#q3
## Escreva uma função que receba duas listas e retorne outra lista com os elementos que estão presentes em apenas uma das listas.

def obter_lista(nome_da_lista):
    while True:
        entrada = input(f"digite os elementos da {nome_da_lista} separados por vírgula (ex: 1,2,3): ")
        try:
            lista = [int(item.strip()) for item in entrada.split(',')]
            return lista
        except ValueError:
            lista = [item.strip() for item in entrada.split(',')]
            return lista

def elementos_exclusivos(lista1, lista2):
    set1 = set(lista1)
    set2 = set(lista2)

    exclusivos1 = set1 - set2
    exclusivos2 = set2 - set1

    resultado = list(exclusivos1.union(exclusivos2))

    return resultado

lista_usuario_1 = obter_lista("primeira lista")
lista_usuario_2 = obter_lista("segunda lista")

resultado_exclusivo_usuario = elementos_exclusivos(lista_usuario_1, lista_usuario_2)
print(f"Lista 1: {lista_usuario_1}")
print(f"Lista 2: {lista_usuario_2}")
print(f"Elementos presentes em apenas uma das listas: {resultado_exclusivo_usuario}")
```

```python
#q4

#Dada uma lista de números inteiros, escreva uma função para encontrar o segundo maior valor na lista.

def obter_lista():
    entrada = input("Digite os elementos da lista separados por vírgula (ex: 1,2,3): ")
    lista = [int(item.strip()) for item in entrada.split(',')]
    lista.sort(reverse=True)
    print("Segundo maior valor:", lista[1])
    return lista[1]
obter_lista()
```

```python
#q5
#Crie uma função que receba uma lista de tuplas, cada uma contendo o nome e a idade de uma pessoa, e retorne a lista ordenada pelo nome das pessoas em ordem alfabética

def cadastro_pessoas():
    qtd = int(input("Quantas pessoas você vai cadastrar? "))
    pessoas = []
    i = 1
    while i <= qtd:
        entrada = input(f"Digite o nome e a idade da pessoa {i}, separados por vírgula (ex: Ana,30): ")
        nome, idade = entrada.split(",")
        pessoas.append((nome.strip(), int(idade.strip())))
        i += 1
    pessoas.sort(key=lambda pessoa: pessoa[0])

    print("Lista ordenada:", pessoas)
    return pessoas

cadastro_pessoas()
```

```python
#q6
# Como identificar e tratar outliers em uma coluna numérica usando desvio padrão ou quartis?

import pandas as pd

df = pd.DataFrame({"valores": [10, 12, 13, 15, 100, 14, 13, 12]})

# usando desvio padrão
media = df["valores"].mean()
desvio = df["valores"].std()
limite_inferior = media - 3 * desvio
limite_superior = media + 3 * desvio
outliers_std = df[(df["valores"] < limite_inferior) | (df["valores"] > limite_superior)]

# usando quartis IQR
Q1 = df["valores"].quantile(0.25)
Q3 = df["valores"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers_iqr = df[(df["valores"] < limite_inferior) | (df["valores"] > limite_superior)]

print("Outliers pelo desvio padrão:\n", outliers_std)
print("\nOutliers pelo IQR:\n", outliers_iqr)
```

```python
#q10

# Utilizando pandas, como lidar com valores ausentes (NaN) em um DataFrame?

import pandas as pd
import numpy as np

df = pd.DataFrame({
    "A": [1, 2, np.nan, 4],
    "B": [np.nan, 2, 3, 4]
})

print("Original\n", df)
df_drop = df.dropna()
df_fill = df.fillna(0)
df_mean = df.fillna(df.mean())

print("\nRemovendo NaN:\n", df_drop)
print("Preenchendo com 0:\n", df_fill)
print("Preenchendo com média:\n", df_mean)
```
